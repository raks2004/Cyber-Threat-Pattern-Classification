# OPTIMIZE DATA

```python
print(df.memory_usage(deep=True).sum())

# Step 1: Downcast float64 to float32
float_columns = df.select_dtypes(include=['float64']).columns
df[float_columns] = df[float_columns].apply(pd.to_numeric, downcast='float')

# Step 2: Downcast int64 to int32 if the range allows
int_columns = df.select_dtypes(include=['int64']).columns
for col in int_columns:
    if df[col].min() >= -2147483648 and df[col].max() <= 2147483647:
        df[col] = df[col].astype('int32')

# Step 3: Ensure int8 columns are within range
int8_columns = df.select_dtypes(include=['int8']).columns
for col in int8_columns:
    if df[col].min() >= -128 and df[col].max() <= 127:
        df[col] = df[col].astype('int8')
```

```python
print(df.memory_usage(deep=True).sum())
```

```python
# Flow Byte to Packet Ratio
df['Flow Byte to Packet Ratio'] = df['Flow Bytes/s'] / df['Flow Packets/s']

# Forward and Backward Packet Ratio
df['Fwd to Bwd Packet Ratio'] = df['Fwd Packets/s'] / df['Bwd Packets/s']

# Average IAT per Direction
df['Avg IAT'] = (df['Fwd IAT Mean'] + df['Bwd IAT Mean']) / 2

# Forward and Backward Packet Length Mean Ratio
df['Fwd to Bwd Packet Length Ratio'] = df['Fwd Packet Length Mean'] / df['Bwd Packet Length Mean']

# Flow IAT Total and Mean Ratio
# Updated to use 'Fwd IAT Total' and 'Bwd IAT Total' instead of 'Flow IAT Total'
df['Fwd to Bwd IAT Total Ratio'] = df['Fwd IAT Total'] / df['Bwd IAT Total']

# Active Time to Idle Time Ratio
df['Active to Idle Ratio'] = df['Active Mean'] / df['Idle Mean']

# Forward to Backward Window Size Ratio
df['Fwd to Bwd Win Size Ratio'] = df['Init Fwd Win Bytes'] / df['Init Bwd Win Bytes']

# Fwd/Bwd PSH Flag count ratio (Can be important for certain types of attacks)
df['Fwd to Bwd PSH Flags Ratio'] = df['Fwd PSH Flags'] / (df['SYN Flag Count'] + df['URG Flag Count'] + 1)

# Subflow Forward and Backward Packets Ratio
df['Subflow Fwd to Bwd Packets Ratio'] = df['Subflow Fwd Packets'] / df['Subflow Bwd Packets']

# Subflow Fwd and Bwd Bytes Ratio
df['Subflow Fwd to Bwd Bytes Ratio'] = df['Subflow Fwd Bytes'] / df['Subflow Bwd Bytes']
```

```python
# List of columns to drop
columns_to_drop = [
    'Fwd Packet Length Max', 'Fwd Packet Length Mean', 'Fwd Packet Length Std',
    'Bwd Packet Length Max', 'Bwd Packet Length Mean', 'Bwd Packet Length Std',
    'Fwd IAT Total', 'Fwd IAT Mean', 'Fwd IAT Std', 'Fwd IAT Max', 'Fwd IAT Min',
    'Bwd IAT Total', 'Bwd IAT Mean', 'Bwd IAT Std', 'Bwd IAT Max', 'Bwd IAT Min',
    'Fwd PSH Flags', 'SYN Flag Count', 'URG Flag Count',
    'Fwd Header Length', 'Bwd Header Length',
    'Fwd Packets/s', 'Bwd Packets/s',
    'Packet Length Max', 'Packet Length Mean', 'Packet Length Std', 'Packet Length Variance',
    'Subflow Fwd Packets', 'Subflow Fwd Bytes', 'Subflow Bwd Packets', 'Subflow Bwd Bytes',
    'Init Fwd Win Bytes', 'Init Bwd Win Bytes',
    'Fwd Act Data Packets', 'Fwd Seg Size Min',
    'Active Mean', 'Active Std', 'Active Max', 'Active Min',
    'Idle Mean', 'Idle Std', 'Idle Max', 'Idle Min'
]

# Drop columns
df_reduced = df.drop(columns=columns_to_drop)

# Verify the remaining columns
print("Remaining Columns: ", df_reduced.columns)
print("New Shape: ", df_reduced.shape)
```

```python
# Encoding the 'Label' and 'ClassLabel' columns using Label Encoding (or One-Hot Encoding if preferred)
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode the 'Label' and 'ClassLabel' columns
df['Label'] = label_encoder.fit_transform(df['Label'])
df['ClassLabel'] = label_encoder.fit_transform(df['ClassLabel'])

# Separate features (X) and target (y)
X = df.drop(columns=['Label', 'ClassLabel'])  # Features
y_label = df['Label']  # Target for 'Label'
```

```python
from sklearn.model_selection import train_test_split

# Split data for 'Label' target
X_train, X_test, y_train_label, y_test_label = train_test_split(X, y_label, test_size=0.2, random_state=42)
```

```python
import numpy as np

# Check for infinity values
print("Checking for Infinity in the dataset:")
print((X_train == np.inf).sum())  # Checking for +inf
print((X_train == -np.inf).sum())  # Checking for -inf

# Check for NaN values
print("Checking for NaN in the dataset:")
print(X_train.isnull().sum())  # Check for NaN values

# Check for extremely large values
print("Checking for extremely large values (greater than a threshold):")
print((X_train > 1e10).sum())  # Check for values greater than 10 billion (adjust threshold if needed)
```

```python
# Replace infinity values with the maximum finite value in the dataset
X_train.replace([np.inf, -np.inf], np.nan, inplace=True)

# Replace NaN values with the median of each column
X_train.fillna(X_train.median(), inplace=True)

# Apply the same transformation to X_test
X_test.replace([np.inf, -np.inf], np.nan, inplace=True)
X_test.fillna(X_test.median(), inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# Apply PCA (choose number of components)
pca = PCA(n_components=50, random_state=42)  
X_train_pca = pca.fit_transform(X_scaled)
X_test_pca = pca.transform(X_test)
```

# MODELING

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Initialize RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=30, max_depth=10, max_features='sqrt', n_jobs=-1, random_state=42)

# Fit the RandomForest model
rf_model.fit(X_train_pca, y_train_label)

# Make predictions with the RandomForest model
y_pred_rf = rf_model.predict(X_test_pca)

# Evaluate the RandomForest model
print("Random Forest Accuracy: ", accuracy_score(y_test_label, y_pred_rf))
print(classification_report(y_test_label, y_pred_rf))
```

```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score, classification_report

lgbm_model = lgb.LGBMClassifier(n_estimators=30, max_depth=10, num_leaves=31, max_bin=255, random_state=42)

# Fit the LGBM model
lgbm_model.fit(X_train_pca, y_train_label)

# Make predictions with the LGBM model
y_pred_lgbm = lgbm_model.predict(X_test_pca)

# Evaluate the LGBM model
print("LightGBM Accuracy: ", accuracy_score(y_test_label, y_pred_lgbm))
print(classification_report(y_test_label, y_pred_lgbm))
```

```python
from sklearn.ensemble import VotingClassifier

# Create a VotingClassifier with both models
voting_model = VotingClassifier(estimators=[('rf', rf_model), ('lgbm', lgbm_model)], voting='hard')

# Fit the VotingClassifier on the training data
voting_model.fit(X_train_pca, y_train_label)

# Make predictions with the VotingClassifier
y_pred_voting = voting_model.predict(X_test_pca)

# Evaluate the VotingClassifier
print("Voting Classifier Accuracy: ", accuracy_score(y_test_label, y_pred_voting))
print(classification_report(y_test_label, y_pred_voting))
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

# Perform cross-validation
cv_scores = cross_val_score(voting_model, X_train_pca, y_train_label, cv=5, scoring='accuracy')  # cv=5 for 5-fold cross-validation

# Print the results
print("Cross-validation scores: ", cv_scores)
print(f"Mean cross-validation accuracy: {np.mean(cv_scores):.4f}")
print(f"Standard deviation of cross-validation accuracy: {np.std(cv_scores):.4f}")
```
